```python
# step1 필요한 모듈 import
from bs4 import BeautifulSoup
from selenium import webdriver
import time
import sys
import math
import numpy
import numpy as np
import pandas as pd
import xlwt
import re #정규식
```

```python
path = "chromedriver_240.exe"
driver = webdriver.Chrome(path)

driver.get('https://news.naver.com/main/read.nhn?m_view=1&mode=LSD&mid=shm&sid1=100&oid=088&aid=0000637370')
time.sleep(2)
```

```python
cnt = 35
page_cnt = math.ceil(cnt/10)
```

```python
html = driver.page_source
soup = BeautifulSoup(html,'html.parser')

result = soup.find('div','u_cbox_head').find('span','u_cbox_count')
result2 = result.get_text()

print('='*80)
result3 = result2.replace(",","")
result4 = re.search('\d+',result3) #re.search를 이용한 정규식 활용 result3에서 숫자만 골라내라
search_cnt = int(result4.group())

if cnt > search_cnt:
    cnt = search_cnt
    
print('전체 검색 결과 건수 :', search_cnt,'건')
print('실제 최종 출력 건수 :', cnt,'건')
print('실제 출력될 페이지 수 :', page_cnt)
```

```python
driver.find_element_by_xpath('//*[@id="cbox_module"]/div/div[9]/a/span/span/span[1]').click()
time.sleep(2)
```

```python
reple_result = soup.find('div','u_cbox_content_wrap').find('ul')
slist = reple_result.find_all('li')
```

```python
count = 0   
writer_id2 = []
review2 = []
write_date2 = []
like2 = []
dislike2 = []
for li in slist:
    count += 1
    print('\n')
    print('%s번째 댓글 수집중 ==============='%count)
    
    writer_id = li.find('span','u_cbox_nick').get_text()
    print('1. 작성자ID',writer_id)
    writer_id2.append(writer_id)
    try:
        review = li.find('span','u_cbox_contents').get_text()
        print('2.리뷰: ',review)
    except AttributeError:
        review = '작성자에 의해 삭제된 댓글입니다.'
        print('2.리뷰: ',review)
    review2.append(review)
    
    write_date = li.find('span', class_='u_cbox_date').get_text()
    print('3.작성일자',write_date)
    write_date2.append(write_date)
    
    
    try:
        like = li.find('em','u_cbox_cnt_recomm').get_text()
        dislike = li.find('em','u_cbox_cnt_unrecomm').get_text()
    except AttributeError:
        like = '-'
        dislike = '-'
    print('4.좋아요',like)
    like2.append(like)
    print('5.싫어요',dislike)
    dislike2.append(dislike)
    
    if count == cnt:
        break
```
