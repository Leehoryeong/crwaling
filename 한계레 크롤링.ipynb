```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
import sys
import pandas as pd
import numpy as np
import math
```

```python
path = 'C:\\Users\\01048\\Desktop\\crawling\\chromedriver_240.exe'
driver = webdriver.Chrome(path)
```

```python
url = 'http://www.hani.co.kr/'
driver.get(url)
```

```python
driver.find_element_by_link_text('더보기').click()
driver.find_element_by_link_text('경제일반').click()
```

```python
cnt = int(input('몇개 크롤링 할까요?: '))
pages = math.ceil(cnt/15)
```

```python
titles = []
summaries = []
dates = []
page_num = 1
no = 1

for page in range(1,pages+1):
    html = driver.page_source
    soup = BeautifulSoup(html,'html.parser')
    contents = soup.find('div','section-list-area').find_all('div','article-area')
    page_url = soup.find('div','paginate').find('a').attrs['href']
    mid_url = page_url[0:34]
    end_url = page_url[35:40]
    for content in contents:
        title = content.find('h4','article-title').get_text().replace('\n','')
        summary = content.find('p','article-prologue').find('a').get_text().replace('\n','')
        date = content.find('span','date').get_text()
        titles.append(title)
        summaries.append(summary)
        dates.append(date)
        print('%s페이지 %s번째 게시글 크롤링 중'%(page_num, no))
        no += 1
        if no == cnt:
            break
    page_num += 1
    full_url = url + mid_url + str(page_num) + end_url
    driver.get(full_url)
    time.sleep(1)
```

```python
data = pd.DataFrame({'제목':titles,'내용요약':summaries,'날짜':dates})
```
