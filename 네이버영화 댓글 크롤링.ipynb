```python
from selenium import webdriver
from bs4 import BeautifulSoup
import math
import time
import sys
import numpy as np
import pandas as pd
import xlwt
import urllib.request
import os
```

```python
path = 'chromedriver_240.exe'
driver = webdriver.Chrome(path)
driver.get('https://movie.naver.com/')
```

```python
movie_name = input('검색할 영화이름:')
```

```python
element = driver.find_element_by_id('ipt_tx_srch')
element.send_keys(movie_name)
time.sleep(1)
driver.find_element_by_xpath('//*[@id="jAutoMV"]/ul/li/a/div').click()
driver.find_element_by_xpath('//*[@id="content"]/div[1]/div[4]/div[5]/div[2]/div[1]/a').click()
```

```python
driver.switch_to.frame('pointAfterListIframe') #iframe은 변환해줘야 해!!
```

```python
cnt = int(input('몇개 크롤링 할래요?: '))
page_num = math.ceil(cnt/10)
```

```python
no = 1
for page in range(1,page_num+1):
    print('%s페이지'%page)
    html = driver.page_source
    soup = BeautifulSoup(html,'html.parser')
    total_cnt = soup.find('div','input_netizen').find('div','score_total').find('em').get_text()
    print(total_cnt)
    num = 0
    comments = soup.find('div','input_netizen').find('div','score_result').find('ul').find_all('li')
    for comment in comments:
        score = comment.find('div','star_score').find('em').get_text()

        content = comment.find('div','score_reple').find('p').find('span',id = '_filtered_ment_%s'%num).get_text().replace('\n','')
        num += 1
        try:
            yorn = comment.find('p').find('span','ico_viewer').get_text()
        except AttributeError:
            yorn = '비관람객'

        box = comment.find('dt').find_all('em')
        writer = box[0].get_text().replace('\n','')
        date = box[1].get_text()

        print('='*50)
        print('%s번째 댓글'%no)
        
        print('1.점수 : ',score)
        print('2. 내용 :','<',yorn,'>','\n',content)
        print('3. 작성자: ', writer)
        print('4. 작성일자: ',date)
        
        if no == cnt:
            break
        no += 1
    driver.find_element_by_xpath('//*[@id="pagerTagAnchor2"]').click()

        #yorn = box[1].get_text().replace('\n','')

    
```
