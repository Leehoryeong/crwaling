```python
from selenium import webdriver
from bs4 import BeautifulSoup
import math
import time
import os
import xlwt
import pandas as pd
import numpy as np
```

```python
path = 'C:\\Users\\01048\\Desktop\\crawling\\chromedriver_240.exe'
driver = webdriver.Chrome(path)
```

```python
year = '2019'
mon = ['01','02','03','04','05','06','07','08','09','10','11','12']

start_date2 = []
end_date2 = []

for i in range(0,len(mon)):
    if mon[i] == '02':
        sdate = year + mon[i] + '01'
        start_date2.append(sdate)
        
        edate = year + mon[i] + '28'
        end_date2.append(edate)
    elif mon[i]=='04' or mon[i]=='06' or mon[i]=='09' or mon[i]=='11':
        sdate = year + mon[i] + '01'
        start_date2.append(sdate)
        
        edate = year + mon[i] + '30'
        end_date2.append(edate)
    else:
        sdate = year + mon[i] + '01'
        start_date2.append(sdate)
        
        edate = year + mon[i] + '31'
        end_date2.append(edate)

```

```python
month =int(input('크롤링할 월은 몇월?'))
```

```python
try:
    if not os.path.exists('C:\\Users\\01048\\Desktop\\crawling\\result\\다음랭킹뉴스 2019년 0%s월'%month):
        os.makedirs('C:\\Users\\01048\\Desktop\\crawling\\result\\다음랭킹뉴스 2019년 0%s월'%month)
except OSError:
    print('존재')
    
sexes = ['female','male']
ages = ['20s','30s','40s','50s']
url = 'https://media.daum.net/ranking/age/?regDate='

for i in range(int(start_date2[month-1]),int(end_date2[month-1])+1):
    full_url = url +  str(i)
    driver.get(full_url)
    time.sleep(2)
    html = driver.page_source
    soup = BeautifulSoup(html,'html.parser')
    data = pd.DataFrame()
    no = 1
    for age in ages:
        years = soup.find('div','item_age item_%s'%age)
        for sex in sexes:
            titles = years.find('div','rank_%s'%sex).find('ol').find_all('li')
            print('%s %s'%(age,sex))
            contents = []
            for title in titles:
                title = title.find('a').get_text()
                contents.append(title)
                print(title)
            data['%s대 %s'%(age[0:2],sex)] = contents
            no += 1
            print('='*40)
    data.to_excel('C:\\Users\\01048\\Desktop\\crawling\\result\\다음랭킹뉴스 2019년 0%s월\\'%month + str(i)+ '.xlsx')
```
