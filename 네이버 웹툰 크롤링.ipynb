```python
# step1 필요한 모듈 import
from bs4 import BeautifulSoup
from selenium import webdriver
import time
import sys
import math
import numpy as np
import pandas as pd
import xlwt
import urllib.request
import os
```

```python
now = time.localtime()
week = ( '월요', '화요', '수요', '목요', '금요', '토요', '일요' );
yoil = week[now.tm_wday]
```

```python
path = 'chromedriver_240.exe'
driver = webdriver.Chrome(path)

driver.get('https://www.naver.com/')
time.sleep(1)
driver.find_element_by_xpath('//*[@id="PM_ID_serviceNavi"]/li[9]/a/span[1]').click()
time.sleep(1)
driver.find_element_by_class_name('Ntxt_webtoon').click()
```

```python
html = driver.page_source
soup = BeautifulSoup(html,'html.parser')
webtoons = soup.find('div','list_area daily_all').find_all('div', 'col')
for webtoon in webtoons:
    # 오늘 요일 확인 후 맞는 요일 선택
    if yoil in webtoon.get_text():
        titles = webtoon.find_all('li')
        # 해당 요일의 제목 뽑아오기 
        for title in titles:
            jemok = title.find('a','title').get_text()
            print(jemok)
            #웹툰 접속
            driver.find_element_by_link_text(jemok).click()
            time.sleep(1)
            
            #해당 주제의 첫번째 웹툰 클릭
            html2 = driver.page_source
            soup2 = BeautifulSoup(html2,'html.parser')
            content = soup2.find('tbody').find('td','title').find('a').get_text()
            driver.find_element_by_link_text(content).click()
            
            #사진 링크 다운
            html3 = driver.page_source
            soup3 = BeautifulSoup(html3,'html.parser')
            imgs = soup3.find('div','wt_viewer').find_all('img')
            img_src = []
        
            for img in imgs:
                img_src.append(img['src'])
            
            # 헤더 설정(forbidden 오류 해결 용)
            opener=urllib.request.build_opener()
            opener.addheaders=[('User-Agent','Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/36.0.1941.0 Safari/537.36')]
            urllib.request.install_opener(opener)
            dir_path = 'C:\\Users\\01048\\Desktop\\crawling\\result\\'
            dir_name = jemok
            os.makedirs(dir_path + '/' + dir_name + '/')
            
            # 사진 저장 시작
            
            for i in range(0,len(img_src)):
                cut = 1
                outfolder = dir_path + '/' + dir_name + '/'
                outfile = jemok + content[0:3] + '%s'%cut +'.jpg'
                try:
                    urllib.request.urlretrieve(img_src[i],outfolder + outfile) #urlretrieve 함수: 그 url 주소로 가서 이미지를 다운 받는 함수
                except:
                    continue # 이미지를 찾으러 갖는데 이미지가 없는 경우 넘어가라 
                time.sleep(0.5)
                print("%s 번째 이미지 저장중입니다 =========="%cut)
                cut += 1
            driver.back()
            time.sleep(1)
            driver.back()
            time.sleep(1)
```
